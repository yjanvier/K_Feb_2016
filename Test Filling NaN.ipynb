```python
def Frequency_Encoder(row, frequencies_dict):
    #print (row.shape)
    for i in range (row.shape[0]):
        try :
            value = row.iloc[i]
            row.iloc[i] = frequencies_dict[i].loc[value]
        except:
            #print ('nan')
            pass
    return row
```

```python
def Calculate_Frequencies(df):
    frequencies_dict={}
    for i in range (df.shape[1]):
        serie = df.iloc[:,i].copy()
        values = serie.value_counts()
        sum_values = values.sum()
        frequencies = values/sum_values
        frequencies_dict[i] = frequencies
    return frequencies_dict
```

```python
def Variables_Correlations_Dict (df_wo_nans):
    df_columns = df_wo_nans.columns
    correlation_matrix = np.abs(np.corrcoef(df_wo_nans.T))
    correlation_dict={}
    correlation_matrix[correlation_matrix==1]=0
    for i in range(correlation_matrix.shape[0]) :
        max_to_min_indices = np.argsort(correlation_matrix[::,i])[::-1]
        correlation_dict[i] = (max_to_min_indices,
                               correlation_matrix[max_to_min_indices,i])
    return correlation_dict
```

```python
def Get_Indices_To_Solve (variable_number, df_with_nans, variables_correlations_dict, df_columns, level):
    
    nan_values_starting_variable = pd.isnull(df_with_nans[df_columns[variable_number]])
    nan_indices_starting_variable = nan_values_starting_variable[nan_values_starting_variable].index
    target_variable = variables_correlations_dict[variable_number][0][level]
    nan_values_target_variable = pd.isnull(df_with_nans[df_columns[target_variable]])
    nan_indices_target_variable = nan_values_starting_variable[nan_values_target_variable].index
    indices_to_resolve = set(nan_indices_starting_variable) - set(nan_indices_target_variable)
    return target_variable,indices_to_resolve

def Get_Indices_To_Solve2 (variable_number, target_variable, df_with_nans, df_columns):
    
    nan_values_starting_variable = pd.isnull(df_with_nans[df_columns[variable_number]])
    nan_indices_starting_variable = nan_values_starting_variable[nan_values_starting_variable].index
    nan_values_target_variable = pd.isnull(df_with_nans[df_columns[target_variable]])
    nan_indices_target_variable = nan_values_starting_variable[nan_values_target_variable].index
    indices_to_resolve = set(nan_indices_starting_variable) - set(nan_indices_target_variable)
    return indices_to_resolve

def Get_Solving_Dicts (indices_to_solve_dict, resolution_strategy_dict, df_with_nans, variables_correlations_dict, df_columns, level):
        
    for variable_number in variables_correlations_dict.keys():
        #print ('Level : %d - Variable : %d'%(level,variable_number))
        target_variable,potential_additional_indices_to_solve = Get_Indices_To_Solve (variable_number, df_with_nans, 
                                                                 variables_correlations_dict, df_columns, level)

        if (len(potential_additional_indices_to_solve) >0):

            if variable_number in indices_to_solve_dict.keys():
                #print ('Updating indices')
                current_indices_to_solve = indices_to_solve_dict[variable_number]
                updated_indices_to_solve = potential_additional_indices_to_solve | current_indices_to_solve
                indices_to_solve_dict[variable_number]=updated_indices_to_solve
                additional_indices_to_solve_size = len(updated_indices_to_solve) - len(current_indices_to_solve)
            else :
                #print ('No existing indices - creating a fresh list')
                additional_indices_to_solve_size = len(potential_additional_indices_to_solve)
                indices_to_solve_dict[variable_number] = potential_additional_indices_to_solve              

            if (additional_indices_to_solve_size>0):
                flag='level'+str(-level)
                #print ('Target Variable %d : %d additional indices to solve'%(target_variable,additional_indices_to_solve_size))
                resolution_strategy_dict[(variable_number,target_variable)]=(flag,
                                                                             variables_correlations_dict[variable_number][1][level],
                                                                            additional_indices_to_solve_size)    
    #print (resolution_strategy_dict)
    resolution_strategy_sorted_dict = dict(OrderedDict(sorted(resolution_strategy_dict.items(), key=operator.itemgetter(1), reverse = True)))
    #print (type(resolution_strategy_sorted_dict))
    return resolution_strategy_sorted_dict, indices_to_solve_dict
```

```python
def Cluster_Variables (df_with_nans, n_clusters=15):
    clusters_dict={}
    for i in range(df_with_nans.shape[1]) :
        if (i%10==0):
            print ('Clustering Variable :%d'%i)
        values_to_cluster = df_with_nans.iloc[:,i].dropna().values.reshape(-1, 1)
        clusters_dict[i] = KMeans(n_clusters=n_clusters).fit(values_to_cluster)
        cluster = clusters_dict[i]
        
        df_with_nans.loc[:,df_columns[i]+'_labels']=np.nan
        df_with_nans.loc[pd.notnull(df_with_nans[df_columns[i]]),df_columns[i]+'_labels']=cluster.labels_

    return clusters_dict, df_with_nans    
```

```python
def Filling_Variable (df_with_nans, starting_variable,target_variable,resolution_strategy_dict, indices_to_solve_dict, clusters_dict, 
                      correlation_threshold = 0.5) :
    value = resolution_strategy_dict[(starting_variable,target_variable)]
    level = value[0]
    correlation = resolution_strategy_dict[(starting_variable,target_variable)][1]     
    indices_to_solve = Get_Indices_To_Solve2 (starting_variable, target_variable, df_with_nans, df_columns)
    #print(correlation)
    if (correlation< correlation_threshold):
        print ('Correlation below threshold')
        df_with_nans.loc[indices_to_solve,df_columns[starting_variable]]= df_with_nans.iloc[:,starting_variable].dropna().mean()
        cluster_starting = clusters_dict[starting_variable]
        for indice_to_solve in list(indices_to_solve):
            predicted_label_in_starting_variable = cluster_starting.predict(df_with_nans.loc[indice_to_solve,df_columns[starting_variable]])
            df_with_nans.loc[indice_to_solve, df_columns[starting_variable]+'_labels'] = predicted_label_in_starting_variable
    else :
        
        label_column_name = df_columns[target_variable]+'_labels'
        cluster_starting = clusters_dict[starting_variable]
        cluster_target = clusters_dict[target_variable]

        for indice_to_solve in list(indices_to_solve):
            label_in_target_variable = df_with_nans[label_column_name].loc[indice_to_solve]
            indices_in_target_variables_with_good_label = df_with_nans.loc[df_with_nans[label_column_name]==label_in_target_variable,label_column_name].index
            sliced_df_starting_variable = df_with_nans.loc[indices_in_target_variables_with_good_label, df_columns[starting_variable]]
            sliced_df_starting_variable_labels = df_with_nans.loc[indices_in_target_variables_with_good_label, df_columns[starting_variable]+'_labels']
            mean_value_starting_variable = sliced_df_starting_variable.dropna().mean()
            df_with_nans.loc[indice_to_solve, df_columns[starting_variable]] = mean_value_starting_variable
            predicted_label_in_starting_variable = cluster_starting.predict(mean_value_starting_variable)
            shoud_be_label = sliced_df_starting_variable_labels.value_counts().index[0]
            df_with_nans.loc[indice_to_solve, df_columns[starting_variable]+'_labels'] = predicted_label_in_starting_variable
            #if (predicted_label_in_starting_variable!=shoud_be_label):
                #print('Strange Assignement')
    indices_to_solve_dict[starting_variable] = indices_to_solve_dict[starting_variable] - set(indices_to_solve)
    return df_with_nans, indices_to_solve_dict                                                        
```

```python
def Solving_Matrix (df_with_nans, df_wo_nans, n_clusters=2):
    
     
    print ('Creating Clusters')
    print (df_with_nans.shape)
    clusters_dict, df_with_nans = Cluster_Variables (df_with_nans, n_clusters=n_clusters)
    print ('Clusters created')
    
    initial_nan_values_number = pd.isnull(df_with_nans).sum().sum()
    initial_non_nan_values_number = pd.notnull(df_with_nans).sum().sum()
    
    initial_ratio = initial_nan_values_number*1.0/(initial_nan_values_number+initial_non_nan_values_number)
    print ('NaN ratio : %f'%initial_ratio)
    
    
    print ('Creating Variables Correlations Dict')
    variables_correlations_dict=Variables_Correlations_Dict (df_wo_nans)
    
    indices_to_solve_dict={}
    resolution_strategy_dict={}
    round=0
    samples_filled = 0
    correlations_sum = 0
    for current_level in range (len(variables_correlations_dict.keys())-1):
        print ('Starting Resolution at Level %d'%current_level)
        resolution_strategy_dict, indices_to_solve_dict = Get_Solving_Dicts (indices_to_solve_dict, resolution_strategy_dict,
                                                                                 df_with_nans, variables_correlations_dict, 
                                                                                 df_columns, level=current_level)
        while (len(resolution_strategy_dict.keys())>0 or (round==0)):
            round +=1
            
            resolution_strategy_dict, indices_to_solve_dict = Get_Solving_Dicts (indices_to_solve_dict, resolution_strategy_dict,
                                                                                 df_with_nans, variables_correlations_dict, 
                                                                                 df_columns, level=current_level)
            
            pairs_to_address = resolution_strategy_dict.copy().keys()
            
            for (starting_variable,target_variable) in pairs_to_address :
                
                strategy_dict_value = resolution_strategy_dict[(starting_variable,target_variable)]
                print ('Filling Variables for (%d, %d) : Level %s - %d samples - %f correlation' %(starting_variable,target_variable,
                                                                                      strategy_dict_value[0],
                                                                                      strategy_dict_value[2],
                                                                                      strategy_dict_value[1]))
                #### Correlation Threshold hardcoded!!
                if (strategy_dict_value[1]>=0.5):
                    samples_filled += strategy_dict_value[2]
                    correlations_sum += strategy_dict_value[1]*strategy_dict_value[2]
                
                nan_values_before = pd.isnull(df_with_nans[df_columns[starting_variable]]).sum()
                #print('I am here!')
                df_with_nans,indices_to_solve_dict = Filling_Variable (df_with_nans, starting_variable,
                                                 target_variable,resolution_strategy_dict, indices_to_solve_dict,clusters_dict)
                
                nan_values_after = pd.isnull(df_with_nans[df_columns[starting_variable]]).sum()
                
                resolution_strategy_dict.pop((starting_variable,target_variable), None)
                
                resolution_strategy_dict, indices_to_solve_dict = Get_Solving_Dicts (indices_to_solve_dict, resolution_strategy_dict,
                                                                                 df_with_nans, variables_correlations_dict, 
                                                                                 df_columns, level=current_level)
            nan_values_number = pd.isnull(df_with_nans).sum().sum()
            solving_progress = (initial_nan_values_number-nan_values_number) * 1.0 / initial_nan_values_number
            average_correlation = correlations_sum * 1.0 / samples_filled
            print ('SOLVING PROGRESS Round %d Level %d : %f - Average Correlation : %f'%(round,current_level,solving_progress,
                                                                                        average_correlation))
            if (len(resolution_strategy_dict.keys())==0):
                print ('Resolution Strategy Dict at level %d empty'%current_level)
            df_with_nans.to_csv('../Data/df_with_nans_enriched.csv')
            time.sleep(0.1)
    return df_with_nans
```
