```python
from geopy.geocoders import GoogleV3
import pandas as pd
import time
```

```python
geoloc = pd.read_csv('../../../Geoloc/Address4.csv', sep=',')
```

```python
geoloc.head(2)
```

```python
def geolocalisation (row, geolocator=geolocator):
    time.sleep(1)
    
    full_address = ' '.join([row['ADDRESS_LINE'], row['ZIP CODE'], row['CITY'], row['COUNTRY']])
    zip_address = ' '.join([row['ZIP CODE'], row['CITY'], row['COUNTRY']])
    city_address = ' '.join([row['CITY'], row['COUNTRY']])
    country_address = row['COUNTRY']
    location = geolocator.geocode(full_address)
    if (location is None):
        location = geolocator.geocode(zip_address)
        if (location is None):
            location = geolocator.geocode(city_address)
            if (location is None):
                location = geolocator.geocode(country_address)
                
    print (location.latitude, location.longitude)
```

```python
geolocalisation(geoloc.iloc[0,:], geolocator)
```

```python
geoloc.apply(geolocalisation, axis=1)
```
