```python
for i in range (0, len(dummy_dfs),2):
    pd.concat(dummy_dfs[i:i+1]).to_csv('../Data/dummy_df'+str(i)+str(i+1))
    print (i)
```

```python
for i in range (0, len(dummy_dfs),4):
    df1 = pd.read_csv('../Data/dummy_df'+str(i)+str(i+1))
    df2 = pd.read_csv('../Data/dummy_df'+str(i+2)+str(i+3))
    pd.concat([df1,df2]).to_csv('../Data/dummy_df_new'+str(i)+str(i+3))
    print(i)
```
