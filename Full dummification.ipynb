```python
sample_submission_path = '../Data/sample_submission.csv'
train_path = '../Data/train.csv'
```

```python
pd.read_csv(sample_submission_path).head(2)
```

```python
train_full = pd.read_csv(train_path)
```

```python
train=train_full[:]
```

```python
train.shape
```

```python
pd.DataFrame(pd.isnull(train).sum()).T
```

```python
y=train.target
```

```python
X = train.drop(labels=['target'], axis=1)
```

```python
train.shape
```

```python
X.shape
```

```python
y.shape
```

```python
columns_categorical = X.dtypes[X.dtypes=='object'].index
```

```python
columns_categorical
```

```python
X_categorical = X[columns_categorical]
```

```python
for column in X_categorical.columns:
    X_categorical.loc[pd.notnull(X_categorical[column]),column] = X_categorical.loc[pd.notnull(X_categorical[column]),column].map(lambda cat : column+cat)
```

```python
X_categorical_dummies = pd.get_dummies(X_categorical)
```

```python
X_categorical_dummies.shape
```

```python
X_categorical
```

```python
X_categorical_dummies.head(2)
```

```python
X_categorical_dummies.to_csv('../Data/X_categorical_dummies.csv')
```

```python
columns_numerical = X.dtypes[X.dtypes!='object'].index
```
